## Section 5: Construct the Financial Resilience Index (FRI)

Combine the four pillar sub-scores into a single FRI score (0-100) using both equal and PCA weighting methods.

```python
# Consolidate all indicators
indicators_df = pd.DataFrame(index=df_working.index)

# Economic Resources
indicators_df['liquidity_score'] = economic_df['liquidity_score'].fillna(0)
indicators_df['debt_burden_score'] = economic_df['debt_burden_score'].fillna(0)
indicators_df['housing_affordability_score'] = economic_df['housing_affordability_score'].fillna(0)

# Financial Inclusion
indicators_df['banking_access_score'] = inclusion_df['banking_access_score'].fillna(0)
indicators_df['credit_access_score'] = inclusion_df['credit_access_score'].fillna(0)
indicators_df['investment_access_score'] = inclusion_df['investment_access_score'].fillna(0)
indicators_df['insurance_score'] = inclusion_df['insurance_score'].fillna(0)

# Financial Behavior
indicators_df['employment_stability_score'] = behavior_df['employment_stability_score'].fillna(0)
indicators_df['savings_habit_score'] = behavior_df['savings_habit_score'].fillna(0)
indicators_df['retirement_planning_score'] = behavior_df['retirement_planning_score'].fillna(0)
indicators_df['risk_management_score'] = behavior_df['risk_management_score'].fillna(0)

# Social Capital
indicators_df['household_composition_score'] = social_df['household_composition_score'].fillna(0)
indicators_df['dependents_score'] = social_df['dependents_score'].fillna(0)
indicators_df['geographic_advantage_score'] = social_df['geographic_advantage_score'].fillna(0)

# Calculate pillar sub-scores (EQUAL WEIGHTING)
pillar_scores_equal = pd.DataFrame(index=df_working.index)
pillar_scores_equal['Economic_Resources'] = indicators_df[
    ['liquidity_score', 'debt_burden_score', 'housing_affordability_score']
].mean(axis=1)
pillar_scores_equal['Financial_Inclusion'] = indicators_df[
    ['banking_access_score', 'credit_access_score', 'investment_access_score', 'insurance_score']
].mean(axis=1)
pillar_scores_equal['Financial_Behavior'] = indicators_df[
    ['employment_stability_score', 'savings_habit_score', 'retirement_planning_score', 'risk_management_score']
].mean(axis=1)
pillar_scores_equal['Social_Capital'] = indicators_df[
    ['household_composition_score', 'dependents_score', 'geographic_advantage_score']
].mean(axis=1)

# PCA Analysis
indicators_standardized = StandardScaler().fit_transform(indicators_df)
pca = PCA(n_components=4)
pca_scores = pca.fit_transform(indicators_standardized)
loadings = pca.components_

# Calculate PCA weights for pillars
ec_indicators = ['liquidity_score', 'debt_burden_score', 'housing_affordability_score']
inc_indicators = ['banking_access_score', 'credit_access_score', 'investment_access_score', 'insurance_score']
beh_indicators = ['employment_stability_score', 'savings_habit_score', 'retirement_planning_score', 'risk_management_score']
soc_indicators = ['household_composition_score', 'dependents_score', 'geographic_advantage_score']

indicator_to_idx = {col: i for i, col in enumerate(indicators_df.columns)}

def get_pillar_pca_weight(indicators_list, component_idx=0):
    indices = [indicator_to_idx[ind] for ind in indicators_list]
    return loadings[component_idx, indices].mean()

pca_weights = {
    'Economic_Resources': get_pillar_pca_weight(ec_indicators),
    'Financial_Inclusion': get_pillar_pca_weight(inc_indicators),
    'Financial_Behavior': get_pillar_pca_weight(beh_indicators),
    'Social_Capital': get_pillar_pca_weight(soc_indicators)
}

# Normalize weights
pca_weights = {k: max(0, v) for k, v in pca_weights.items()}
total_weight = sum(pca_weights.values())
pca_weights = {k: v / total_weight if total_weight > 0 else 0.25 for k, v in pca_weights.items()}

pillar_scores_pca = pillar_scores_equal.copy()

print("✓ Pillar Sub-Scores Calculated")
print(f"\nPillar Scores (Equal Weighting):")
print(pillar_scores_equal.describe().round(3))
print(f"\nPCA-Based Pillar Weights:")
for pillar, weight in pca_weights.items():
    print(f"  {pillar}: {weight:.4f}")
```

## Section 3: Build Indicator Engineering Framework

Prepare the data and create individual indicators for each pillar of financial resilience.

```python
import zipfile
import xml.etree.ElementTree as ET
import re

# Define Excel parsing helpers
MAIN_NS = "http://schemas.openxmlformats.org/spreadsheetml/2006/main"
REL_NS = "http://schemas.openxmlformats.org/officeDocument/2006/relationships"
PKG_REL_NS = "http://schemas.openxmlformats.org/package/2006/relationships"
NAMESPACES = {"m": MAIN_NS, "r": REL_NS, "pr": PKG_REL_NS}

def col_letters_to_index(col_letters: str) -> int:
    idx = 0
    for char in col_letters:
        idx = idx * 26 + (ord(char) - ord("A") + 1)
    return idx - 1

def read_shared_strings(zip_file):
    if "xl/sharedStrings.xml" not in zip_file.namelist():
        return []
    root = ET.fromstring(zip_file.read("xl/sharedStrings.xml"))
    out = []
    for shared_item in root.findall("m:si", NAMESPACES):
        text_nodes = shared_item.findall(".//m:t", NAMESPACES)
        out.append("".join(node.text or "" for node in text_nodes))
    return out

def workbook_sheet_paths(zip_file):
    workbook = ET.fromstring(zip_file.read("xl/workbook.xml"))
    workbook_rels = ET.fromstring(zip_file.read("xl/_rels/workbook.xml.rels"))
    rel_id_to_target = {}
    for rel in workbook_rels.findall("pr:Relationship", NAMESPACES):
        rel_id = rel.attrib.get("Id")
        target = rel.attrib.get("Target", "")
        if not rel_id:
            continue
        if target.startswith("/"):
            target = target.lstrip("/")
        elif not target.startswith("xl/"):
            target = f"xl/{target}"
        rel_id_to_target[rel_id] = target
    sheet_paths = {}
    for sheet in workbook.findall("m:sheets/m:sheet", NAMESPACES):
        name = sheet.attrib.get("name")
        rel_id = sheet.attrib.get(f"{{{REL_NS}}}id")
        if not name or not rel_id:
            continue
        target = rel_id_to_target.get(rel_id)
        if target:
            sheet_paths[name] = target
    return sheet_paths

def decode_cell_value(cell, shared_strings):
    cell_type = cell.attrib.get("t")
    value_node = cell.find("m:v", NAMESPACES)
    if cell_type == "s":
        if value_node is None or value_node.text is None:
            return ""
        return shared_strings[int(value_node.text)]
    if cell_type == "inlineStr":
        text_node = cell.find("m:is/m:t", NAMESPACES)
        return text_node.text if text_node is not None and text_node.text else ""
    if cell_type == "b":
        if value_node is None:
            return ""
        return "1" if value_node.text == "1" else "0"
    return value_node.text if value_node is not None and value_node.text is not None else ""

def parse_sheet(zip_file, sheet_path, shared_strings):
    root = ET.fromstring(zip_file.read(sheet_path))
    sheet_data = root.find("m:sheetData", NAMESPACES)
    if sheet_data is None:
        return pd.DataFrame()
    
    sparse_rows = []
    max_col = -1
    
    for row in sheet_data.findall("m:row", NAMESPACES):
        sparse_row = {}
        for cell in row.findall("m:c", NAMESPACES):
            cell_ref = cell.attrib.get("r", "")
            match = re.match(r"([A-Z]+)", cell_ref)
            if not match:
                continue
            col_idx = col_letters_to_index(match.group(1))
            max_col = max(max_col, col_idx)
            sparse_row[col_idx] = decode_cell_value(cell, shared_strings)
        sparse_rows.append(sparse_row)
    
    if max_col < 0 or not sparse_rows:
        return pd.DataFrame()
    
    matrix = []
    for sparse_row in sparse_rows:
        matrix.append([sparse_row.get(i, "") for i in range(max_col + 1)])
    
    header = matrix[0]
    data = matrix[1:]
    columns = []
    for i, name in enumerate(header):
        clean_name = str(name).strip()
        if not clean_name:
            clean_name = f"col_{i}"
        columns.append(clean_name)
    
    return pd.DataFrame(data, columns=columns)

# Load data
data_path = Path('analysis') / 'personal_finance_dataset.xlsx'

try:
    with zipfile.ZipFile(data_path) as zip_file:
        shared_strings = read_shared_strings(zip_file)
        sheet_paths = workbook_sheet_paths(zip_file)
        
        data_sheet = "datathon_finance"
        if data_sheet not in sheet_paths:
            all_sheet_names = list(sheet_paths.keys())
            if len(all_sheet_names) >= 2:
                data_sheet = all_sheet_names[1]
        
        df = parse_sheet(zip_file, sheet_paths[data_sheet], shared_strings)
    
    print(f"✓ Dataset loaded successfully!")
    print(f"  Shape: {df.shape}")
    
except Exception as e:
    print(f"Creating sample dataset (Excel load error: {e})...")
    np.random.seed(42)
    n_samples = 1000
    
    df = pd.DataFrame({
        'PEFATINC': np.random.lognormal(10, 1, n_samples),
        'PWAPRVAL': np.random.lognormal(11, 1.5, n_samples),
        'PWASTDEP': np.random.lognormal(8, 2, n_samples),
        'PWATFS': np.random.lognormal(7, 2, n_samples),
        'PWDPRMOR': np.random.lognormal(9, 1.5, n_samples),
        'PWDSLOAN': np.random.lognormal(6, 2, n_samples),
        'PWDSTCRD': np.random.lognormal(7, 2, n_samples),
        'PWDSTLOC': np.random.lognormal(8, 2, n_samples),
        'PWNETWPG': np.random.lognormal(10, 1.5, n_samples),
        'Banking_Access': np.random.choice([0, 1], n_samples, p=[0.1, 0.9]),
        'Credit_Access': np.random.choice([0, 1], n_samples, p=[0.2, 0.8]),
        'RRSP_Balance': np.random.lognormal(7, 2, n_samples),
        'Insurance_Owned': np.random.choice([0, 1], n_samples, p=[0.3, 0.7]),
        'Employment_Type': np.random.choice(['Full-time', 'Part-time', 'Self-employed'], n_samples),
        'Household_Size': np.random.choice([1, 2, 3, 4, 5], n_samples, p=[0.2, 0.3, 0.25, 0.15, 0.1]),
        'Num_Dependents': np.random.choice([0, 1, 2, 3], n_samples, p=[0.4, 0.3, 0.2, 0.1]),
        'Region': np.random.choice(['Toronto', 'Rest of Ontario'], n_samples, p=[0.4, 0.6]),
        'Num_Income_Earners': np.random.choice([1, 2], n_samples, p=[0.4, 0.6]),
    })
    print(f"✓ Sample dataset created!")
    print(f"  Shape: {df.shape}")

print(f"\nDataset Info:")
print(f"  Columns: {df.shape[1]}")
print(f"  First row: {df.iloc[0].to_dict()}")
```

# Financial Resilience Index (FRI) Analysis
## A Comprehensive Framework for Household Financial Health

### Overview
This notebook constructs a comprehensive **Financial Resilience Index (FRI)** for personal finance analysis. The FRI is a multi-dimensional construct that captures a household's ability to withstand financial shocks and pursue long-term stability.

### The Four Pillars of Financial Resilience
1. **Economic Resources**: Tangible financial buffers (liquidity, debt burden, housing affordability)
2. **Financial Inclusion**: Access to and use of mainstream financial products (banking, credit, investments, insurance)
3. **Financial Behavior**: Actions and planning habits (employment stability, savings, retirement planning, risk management)
4. **Social Capital**: Informal support networks (household composition, dependents, geographic context)

### Methodology
- Extract relevant variables from the SFS dataset
- Assign binary scores (0 = Vulnerable, 1 = Resilient) based on thresholds
- Aggregate to pillar sub-scores (0-1 range)
- Construct final FRI (0-100) using equal weighting and PCA weighting
- Segment households into resilience categories (At-Risk, Coping, Thriving)

### Notebook Structure
1. **Setup**: Import libraries
2. **Data Loading**: Load and explore dataset
3. **Indicator Engineering**: Build indicators for each pillar
4. **Pillar Calculation**: Aggregate indicators to pillar sub-scores
5. **Index Construction**: Create final FRI
6. **Segmentation**: Classify households by resilience
7. **Analysis & Visualization**: Comprehensive insights and charts
8. **Stakeholder Insights**: Actionable recommendations for decision-makers
